1. What is a random variable in probability theory

 Core Idea:

Discrete Random Variables: These can only take on a countable number of values (e.g., the number of heads in a series of coin flips).

4. What are probability distribution functions (PDF)?


- When discussing probability, it's important to distinguish between discrete and continuous random variables. With that in mind, here's a breakdown of probability distribution functions:

 Understanding Probability Distributions

 A probability distribution is a mathematical description of the probabilities of possible values of a random variable. In simpler terms, it tells you how likely it is for a random variable to take on certain values.   
Probability Density Functions (PDFs)

 PDFs specifically apply to continuous random variables.   
A PDF, denoted as f(x), represents the relative likelihood that a continuous random variable will take on a given value.   
Key points about PDFs:  
The value of the PDF at a particular point doesn't represent the actual probability of that exact value (because for continuous variables, that probability is essentially zero).  
Instead, the PDF indicates the "density" of probability around that point. Higher PDF values mean that values in that vicinity are more likely.   
To find the probability that a continuous random variable falls within a certain interval, you calculate the area under the PDF curve over that interval.   
The total area under the PDF curve over the entire range of possible values is always equal to 1.

7. What are the key properties of a Bernoulli distribution?

- The Bernoulli distribution is a fundamental concept in probability theory, describing the probability of a single trial with two possible outcomes. Here are its key properties:   

 Binary Outcomes:  
The defining characteristic of a Bernoulli distribution is that it models a random experiment with only two possible outcomes. These are often labeled as "success" and "failure," or represented numerically as 1 and 0.   
Single Parameter:  
The Bernoulli distribution is governed by a single parameter, denoted as "p."   This parameter represents the probability of "success." Consequently, the probability of "failure" is 1 - p.   
Probability Mass Function (PMF):  
The probability mass function of a Bernoulli random variable (X) is:  
P(X = 1) = p (probability of success)  
P(X = 0) = 1 - p (probability of failure)   
Mean (Expected Value):  
The mean (or expected value) of a Bernoulli distribution is equal to the probability of success: E(X) = p.   
Variance:  
The variance of a Bernoulli distribution is given by: Var(X) = p(1 - p).   
Relationship to Binomial Distribution:  
The Bernoulli distribution is a special case of the binomial distribution, where the number of trials is equal to 1. In other words, a binomial distribution describes the number of successes in multiple independent Bernoulli trials.

11. What are the characteristics of a normal distribution?

- The normal distribution, also known as the Gaussian distribution, is one of the most important probability distributions in statistics. It's characterized by its bell-shaped curve and has several key properties:   

 Bell-Shaped and Symmetrical:  
The graph of the normal distribution is a symmetrical, bell-shaped curve. This means that the data is evenly distributed around the mean.   
Mean, Median, and Mode are Equal:  
In a perfect normal distribution, the mean (average), median (middle value), and mode (most frequent value) are all the same.  
Defined by Two Parameters:  
The normal distribution is fully defined by two parameters:  
Mean (μ): This determines the center of the distribution.   
Standard Deviation (σ): This determines the spread or width of the distribution.  
Empirical Rule (68-95-99.7 Rule):  
This rule states that:  
Approximately 68% of the data falls within one standard deviation of the mean (μ ± 1σ).  
Approximately 95% of the data falls within two standard deviations of the mean (μ ± 2σ).  
Approximately 99.7% of the data falls within three standard deviations of the mean (μ ± 3σ).  
Asymptotic Tails:  
The tails of the normal distribution extend infinitely in both directions, approaching but never touching the horizontal axis.   
Continuous Distribution:  
The normal distribution is a continuous probability distribution, meaning that it can take on any value within its range.   
Probability Density Function (PDF):  
The PDF of a normal distribution is a specific mathematical function that describes the shape of the bell curve.   
These characteristics make the normal distribution a powerful tool for modeling and analyzing a wide range of data in various fields.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

- The Central Limit Theorem (CLT) is a cornerstone of probability theory and statistics. It essentially states that under certain conditions, the distribution of the sample mean approaches a normal distribution, regardless of the shape of the original population distribution. Here's a breakdown:   

 Core Concept:  

 Sampling Distribution of the Mean:  
Imagine repeatedly taking samples of a fixed size from a population and calculating the mean of each sample. The distribution of these sample means is called the "sampling distribution of the mean."   
The Theorem:  
The CLT states that as the sample size increases, the sampling distribution of the mean will become approximately normally distributed, even if the original population is not normally distributed.   
Conditions:  
Typically, the sample size needs to be "sufficiently large" (often considered n ≥ 30).   
The samples must be independent and randomly selected.   
The population must have a finite variance.   
Why It's Critical:  

 Statistical Inference:  
The CLT allows us to make inferences about population parameters (like the mean) using sample statistics. This is fundamental for hypothesis testing and confidence interval estimation.   
Simplified Calculations:  
Because the sampling distribution of the mean becomes normal, we can use the properties of the normal distribution to calculate probabilities and make predictions, even when we don't know the original population's distribution.  
Real-World Applications:  
Many real-world phenomena involve the sum or average of multiple independent factors. The CLT explains why the normal distribution appears so frequently in these situations.  
Foundation for Many Statistical Tests:  
Many statistical tests rely on the assumption of normality. The CLT provides the justification for using these tests, even when the underlying population is not normally distributed.

15. What is the application of Z statistics in hypothesis testing?

- The Z-statistic plays a crucial role in hypothesis testing, particularly when dealing with data that follows a normal distribution or when sample sizes are large. Here's how it's applied:   

 Core Function:  

 The Z-statistic (or Z-score) is used to determine how many standard deviations a sample mean is away from the population mean. This allows us to assess whether the difference between the sample mean and the population mean is statistically significant.   
Applications in Hypothesis Testing:  

 Comparing Sample Means to Population Means:  
A primary application is to test whether the mean of a sample is significantly different from a known or hypothesized population mean.   
This is done by calculating the Z-statistic and comparing it to critical values from the standard normal distribution.  
Comparing Two Sample Means:  
Z-tests can also be used to compare the means of two independent samples, determining if there's a statistically significant difference between them.   
Conditions for Use:  
Z-tests are most appropriate when:  
The population standard deviation is known.   
The sample size is large (typically n ≥ 30), allowing the Central Limit Theorem to apply.   
The data follows a normal distribution.   
Hypothesis Testing Steps:  
In hypothesis testing, the Z-statistic is used in the following general steps:
Formulate hypotheses: Define the null hypothesis (H0) and the alternative hypothesis (H1).  
Calculate the Z-statistic: Using the sample mean, population mean, population standard deviation, and sample size.   
Determine the critical value: Based on the chosen significance level (alpha) and the type of test (one-tailed or two-tailed).  
Make a decision: Compare the calculated Z-statistic to the critical value. If the Z-statistic falls within the rejection region, the null hypothesis is rejected.   
P-values:  
The Z-statistic is also used to calculate p-values, which represent the probability of obtaining the observed results if the null hypothesis were true.
Comparing the p-value to the significance level helps in making decisions about rejecting or failing to reject the null hypothesis.

16. How do you calculate a Z-score, and what does it represent?

- The Z-score is a fundamental statistical measure that helps us understand how a particular data point relates to the rest of a dataset. Here's a breakdown of how to calculate it and what it represents:   

 Calculation:

 The formula for calculating a Z-score is:

 Z = (x - μ) / σ   
Where:  

 Z: The Z-score.   
x: The individual data point.   
μ (mu): The population mean.   
σ (sigma): The population standard deviation.  
If you are working with a sample, and do not have the population standard deviation, you would use the samples standard deviation represented by the letter "s". The formula would then be.   

 Z = (x - x̄) / s  
Where:  

 x̄: The sample mean.   
s: The sample standard deviation.   
What it Represents:  

 A Z-score tells you how many standard deviations a data point is away from the mean of a distribution.   
Positive Z-score: Indicates that the data point is above the mean.   
Negative Z-score: Indicates that the data point is below the mean.   
Z-score of 0: Indicates that the data point is equal to the mean.   
Magnitude of the Z-score: The larger the absolute value of the Z-score, the farther the data point is from the mean, and therefore, the more unusual it is.

 i. Point Estimates:

 Definition:
A point estimate is a single value that is used to estimate an unknown population parameter.   
It's essentially our "best guess" for the parameter.   
Examples:

Simple and easy to calculate.  
Provides a single, specific value.   
However, it doesn't convey any information about the precision or reliability of the estimate.

It gives us a sense of the uncertainty associated with our estimate.

Confidence intervals are the most common type of interval estimate.   
A 95% confidence interval for the population mean would provide a range of values that we are 95% confident contains the true population mean.   
Characteristics:  
Provides a range of plausible values.   
Accounts for the variability of the sample data.   
Indicates the precision of the estimate.  
They are far more useful than point estimates, because they provide a range of values, and a level of certainty.

19. What is the relationship between a Z-score and a confidence interval?

- The relationship between a Z-score and a confidence interval is fundamentally about using the properties of the standard normal distribution to construct a range of plausible values for a population parameter. Here's a breakdown:  

 Z-scores and the Standard Normal Distribution:

 Z-scores represent the number of standard deviations a data point is away from the mean of a standard normal distribution (mean = 0, standard deviation = 1).  
The standard normal distribution is crucial because we can use it to calculate probabilities associated with different Z-scores.   
Confidence Intervals and Z-scores:  

 When constructing a confidence interval for a population mean (especially when the population standard deviation is known or the sample size is large), we use Z-scores to determine the margin of error.  
The margin of error is the range of values added and subtracted from the sample mean to create the confidence interval.  
To create that margin of error, we use the Z-score that corresponds to a desired confidence level. For example:  
For a 95% confidence interval, we use a Z-score of approximately 1.96.   
For a 99% confidence interval, we use a Z-score of approximately 2.576.   
This Z-score represents how many standard errors we need to extend from the sample mean to capture the desired level of confidence.  
The standard error is an estimation of the standard deviation of the sampling distribution of the mean.   
The formula for a confidence interval involving the Z score is:  
Confidence Interval = sample mean ± (Z-score * standard error)  
In essence:  

 Z-scores allow us to quantify how far away from the mean a value is in terms of standard deviations.   
In the context of confidence intervals, Z-scores help us determine how wide the interval needs to be to capture a certain level of confidence that the true population parameter lies within it.  
Therefore, Z scores, and the standard normal distribution, are the base from which the margin of error is calculated, which is then used to construct the confidence interval.

21. What are the assumptions for applying the Central Limit Theorem?

- The Central Limit Theorem (CLT) is a powerful tool, but it relies on certain assumptions to hold true. Here's a breakdown of the key assumptions:   

 i. Independence:

 The samples drawn from the population must be independent of each other. This means that the outcome of one sample should not influence the outcome of any other sample.   
This assumption is crucial for ensuring that the sample means are not biased or correlated.   
ii. Random Sampling:  

 The samples must be selected randomly from the population. This ensures that each member of the population has an equal chance of being included in the sample.   
Random sampling helps to minimize bias and ensures that the samples are representative of the population.    
iii. Finite Variance:  

 The population from which the samples are drawn must have a finite variance. This means that the spread of the data in the population should not be infinitely large.   
If the variance is infinite, the CLT may not hold, and the sampling distribution of the mean may not converge to a normal distribution.   
iv. Sample Size (Typically n ≥ 30):  

 While not a strict assumption, the CLT works best with sufficiently large sample sizes.   
A common rule of thumb is that a sample size of 30 or more is generally considered large enough for the CLT to apply.   
However, if the population is already normally distributed, the CLT can hold even with smaller sample sizes.  
If the population is highly skewed, a larger sample size may be needed for the sampling distribution of the mean to approach normality.  
Important Considerations:  

 Even if these assumptions are not perfectly met, the CLT can still provide a reasonable approximation in many real-world situations.   
However, it's essential to be aware of the potential limitations and to consider the specific context of the data.  
When the population is heavily skewed, or the sample size is very small, other methods may be needed.  

22. What is the concept of expected value in a probability distribution?

- The concept of expected value is fundamental in probability theory and provides a way to understand the "average" outcome of a random variable. Here's a breakdown:   

 Core Idea:

 The expected value of a random variable is the weighted average of all possible values that the variable can take, where the weights are the probabilities of those values occurring.   
In simpler terms, it's the long-term average outcome you would expect if you repeated a random experiment many times.  
Formal Definitions:  

 Discrete Random Variables:  
If X is a discrete random variable with possible values x1, x2, ..., xn and corresponding probabilities P(x1), P(x2), ..., P(xn), then the expected value of X, denoted as E(X), is:  
E(X) = Σ [xi * P(xi)]  
This means you multiply each possible value by its probability and then sum up all those products.   
Continuous Random Variables:  
If X is a continuous random variable with probability density function f(x), then the expected value of X is:  
E(X) = ∫ [-∞, ∞] x * f(x) dx   
This means you integrate the product of each possible value and its probability density over the entire range of possible values.  
Intuitive Explanation:  

 Think of it as the "center of mass" of a probability distribution. It represents the value around which the outcomes of a random experiment tend to cluster in the long run.  
For example, when you roll a fair six-sided die, the expected value is 3.5.  
 This doesn't mean you'll ever roll a 3.5, but it means that if you roll the die many times, the average of your rolls will approach 3.5.   
Importance:  

 The expected value is essential for decision-making in situations involving uncertainty.   
It's used in various fields, including:  
Finance (to evaluate the potential returns of investments)   
Insurance (to calculate premiums)  
Gambling (to assess the fairness of games)   
And many other areas where probabilities are involved.

23. How does a probability distribution relate to the expected outcome of a random variable?

- A probability distribution and the expected outcome (or expected value) of a random variable are closely related. Here's how they connect:   

 Probability Distribution:

 A probability distribution provides a complete description of the possible values a random variable can take and the probabilities associated with each of those values.   
For a discrete random variable, it lists each possible value and its probability.   
For a continuous random variable, it's represented by a probability density function (PDF), which describes the relative likelihood of values within a range.   
Expected Outcome (Expected Value):  

 The expected outcome (or expected value) is a single number that represents the average value you would expect the random variable to take over many repetitions of the random experiment.   
It's a weighted average, where each possible value is weighted by its probability.  
The Relationship:  

 The probability distribution is the foundation for calculating the expected outcome.   
The expected value is derived directly from the probability distribution.  
Specifically, to compute the expected value:  
For discrete variables, you multiply each possible value by its corresponding probability and then sum up the results.   
For continuous variables, you integrate the product of each possible value and its probability density function over the entire range of possible values.  
Therefore, the probability distribution provides the necessary information (possible values and their probabilities) to determine the expected outcome.   
The expected value, in turn, summarizes the "center" or "average" of the probability distribution. It gives you a sense of what to expect on average when the random variable is observed repeatedly.  
The probability distribution shows all possible outcomes, and how likely each is, while the expected value condenses that information to a single, average value.  
In essence, the probability distribution is the "blueprint," and the expected value is a key "summary statistic" derived from that blueprint.



```python
#1. Write a Python program to generate a random variable and display its value.

import random

def generate_and_display_random_variable(distribution_type, **kwargs):
    """
    Generates a random variable based on the specified distribution and displays its value.

    Args:
        distribution_type (str): The type of distribution to use.
                                 Options: 'uniform', 'normal', 'randint', 'choice', etc.
        **kwargs: Keyword arguments specifying the parameters of the distribution.
    """

    try:
        if distribution_type == 'uniform':
            a = kwargs.get('a', 0)  # Default lower bound is 0
            b = kwargs.get('b', 1)  # Default upper bound is 1
            random_value = random.uniform(a, b)
            print(f"Generated Uniform Random Variable: {random_value}")

        elif distribution_type == 'normal':
            mu = kwargs.get('mu', 0)  # Default mean is 0
            sigma = kwargs.get('sigma', 1)  # Default standard deviation is 1
            random_value = random.normalvariate(mu, sigma)
            print(f"Generated Normal Random Variable: {random_value}")

        elif distribution_type == 'randint':
            a = kwargs.get('a', 1) # Default lower bound is 1
            b = kwargs.get('b', 10) # default upper bound is 10
            random_value = random.randint(a,b)
            print(f"Generated Random Integer: {random_value}")

        elif distribution_type == 'choice':
            seq = kwargs.get('seq')
            if not seq:
                print("Error: sequence must be provided for choice.")
                return
            random_value = random.choice(seq)
            print(f"Generated Random Choice: {random_value}")

        else:
            print("Error: Invalid distribution type.")

    except ValueError as e:
        print(f"Error: {e}")
    except TypeError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage:
generate_and_display_random_variable('uniform', a=10, b=20)
generate_and_display_random_variable('normal', mu=5, sigma=2)
generate_and_display_random_variable('randint', a = 1, b = 6)
generate_and_display_random_variable('choice', seq = ["apple", "banana", "cherry"])
generate_and_display_random_variable('invalid')
generate_and_display_random_variable('uniform', a = "invalid", b = 20)
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculates the probability mass function (PMF) of a Bernoulli distribution.

    Args:
        x (int): The value to calculate the probability for (0 or 1).
        p (float): The probability of success (1).

    Returns:
        float: The probability of x.
        None: If x is not 0 or 1, or if p is outside the valid range.
    """

    if not isinstance(x, int) or x not in (0, 1):
        print("Error: x must be either 0 or 1.")
        return None

    if not isinstance(p, (int, float)) or not 0 <= p <= 1:
        print("Error: p must be a float or int between 0 and 1.")
        return None

    if x == 1:
        return p
    else:  # x == 0
        return 1 - p

# Example usage:
p_success = 0.6
prob_0 = bernoulli_pdf(0, p_success)
prob_1 = bernoulli_pdf(1, p_success)

print(f"Probability of x=0: {prob_0}")
print(f"Probability of x=1: {prob_1}")

#Example of invalid inputs.
print(bernoulli_pdf(2, 0.5))
print(bernoulli_pdf(1, 1.5))
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

import matplotlib.pyplot as plt
import numpy as np

def discrete_uniform_cdf(a, b):
    """
    Calculates and plots the CDF of a discrete uniform distribution.

    Args:
        a (int): The lower bound of the distribution.
        b (int): The upper bound of the distribution.
    """

    if not isinstance(a, int) or not isinstance(b, int):
        raise ValueError("a and b must be integers.")

    if a > b:
        raise ValueError("a must be less than or equal to b.")

    values = np.arange(a, b + 1)
    probabilities = np.full(len(values), 1 / (b - a + 1))
    cdf = np.cumsum(probabilities)

    plt.step(values, cdf, where='post')  # Use step plot for CDF
    plt.xticks(values)
    plt.xlabel("Values")
    plt.ylabel("Cumulative Probability")
    plt.title(f"Discrete Uniform Distribution CDF (a={a}, b={b})")
    plt.grid(True)
    plt.show()

# Example usage:
discrete_uniform_cdf(2, 7)
```

```python
#10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import matplotlib.pyplot as plt
import numpy as np
import random

def clt_simulation(population, sample_size, num_samples):
    """
    Simulates the Central Limit Theorem (CLT) for a given population.

    Args:
        population (list or numpy array): The population data.
        sample_size (int): The size of each sample.
        num_samples (int): The number of samples to generate.

    Returns:
        numpy array: An array of sample means.
    """

    if sample_size <= 0 or num_samples <= 0:
        raise ValueError("Sample size and number of samples must be positive integers.")

    sample_means = []
    for _ in range(num_samples):
        sample = random.sample(list(population), sample_size) # random.sample returns a list, so convert population to list.
        sample_means.append(np.mean(sample))

    return np.array(sample_means)

def visualize_clt(population, sample_size, num_samples):
    """
    Visualizes the CLT simulation results.

    Args:
        population (list or numpy array): The population data.
        sample_size (int): The size of each sample.
        num_samples (int): The number of samples to generate.
    """

    sample_means = clt_simulation(population, sample_size, num_samples)

    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black') #Added edgecolor for better visualization
    plt.xlabel("Sample Means")
    plt.ylabel("Frequency")
    plt.title(f"CLT Simulation (Sample Size={sample_size}, Samples={num_samples})")
    plt.grid(True)
    plt.show()

# Example using a non-normal distribution (exponential):
population = np.random.exponential(scale=2, size=10000)  # Exponential distribution

visualize_clt(population, sample_size=30, num_samples=1000)
visualize_clt(population, sample_size=100, num_samples=1000) #Show how increasing the sample size causes the distribution to become more normal.
```

```python
#11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

import matplotlib.pyplot as plt
import numpy as np

def verify_clt_normal(population_mean, population_std, sample_size, num_samples):
    """
    Simulates multiple samples from a normal distribution and verifies the CLT.

    Args:
        population_mean (float): The mean of the normal distribution.
        population_std (float): The standard deviation of the normal distribution.
        sample_size (int): The size of each sample.
        num_samples (int): The number of samples to generate.
    """

    sample_means = []
    for _ in range(num_samples):
        sample = np.random.normal(population_mean, population_std, sample_size)
        sample_means.append(np.mean(sample))

    sample_means = np.array(sample_means)

    # Plot the histogram of sample means
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black') # Added edgecolor for better visualization
    plt.xlabel("Sample Means")
    plt.ylabel("Probability Density")
    plt.title(f"CLT Verification (Normal Pop., n={sample_size}, Samples={num_samples})")
    plt.grid(True)
    plt.show()

    # Calculate the mean and standard deviation of the sample means
    mean_of_means = np.mean(sample_means)
    std_of_means = np.std(sample_means)

    # Theoretical standard deviation of sample means (standard error)
    theoretical_std = population_std / np.sqrt(sample_size)

    print(f"Mean of Sample Means: {mean_of_means}")
    print(f"Standard Deviation of Sample Means: {std_of_means}")
    print(f"Theoretical Standard Error: {theoretical_std}")

# Example usage:
population_mean = 50
population_std = 10
sample_size = 30
num_samples = 1000

verify_clt_normal(population_mean, population_std, sample_size, num_samples)

sample_size = 100 #Show how increasing sample size causes the standard deviation of sample means to decrease, and get closer to the theoretical standard error.
verify_clt_normal(population_mean, population_std, sample_size, num_samples)
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

def z_test_hypothesis(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Performs a Z-test for hypothesis testing.

    Args:
        sample_data (list or numpy array): The sample dataset.
        population_mean (float): The hypothesized population mean.
        population_std (float): The population standard deviation.
        alpha (float): The significance level (default: 0.05).
        alternative (str): The type of alternative hypothesis ('two-sided', 'less', 'greater').
    """

    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std / np.sqrt(sample_size)

    z_statistic = (sample_mean - population_mean) / standard_error

    if alternative == 'two-sided':
        critical_value = stats.norm.ppf(1 - alpha / 2)
        p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
        print("Two-sided test")

    elif alternative == 'less':
        critical_value = stats.norm.ppf(alpha)
        p_value = stats.norm.cdf(z_statistic)
        print("Less than test")

    elif alternative == 'greater':
        critical_value = stats.norm.ppf(1 - alpha)
        p_value = 1 - stats.norm.cdf(z_statistic)
        print("Greater than test")

    else:
        raise ValueError("Invalid alternative hypothesis. Use 'two-sided', 'less', or 'greater'.")

    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"Critical Value: {critical_value:.4f}")
    print(f"P-value: {p_value:.4f}")

    if abs(z_statistic) > critical_value and p_value < alpha:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

# Example usage:
sample_data = [65, 68, 70, 72, 75, 63, 66, 69, 71, 73]
population_mean = 67
population_std = 5

z_test_hypothesis(sample_data, population_mean, population_std)  # Two-sided test
z_test_hypothesis(sample_data, population_mean, population_std, alternative='greater')  # One-sided (greater) test
z_test_hypothesis(sample_data, population_mean, population_std, alternative='less') # One-sided (less) test.

#Example of invalid input.
#z_test_hypothesis(sample_data, population_mean, population_std, alternative = "invalid")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculates a confidence interval for a dataset.

    Args:
        data (list or numpy array): The dataset.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95% confidence).

    Returns:
        tuple: A tuple containing the lower and upper bounds of the confidence interval.
    """

    data_array = np.array(data)
    sample_mean = np.mean(data_array)
    sample_std = np.std(data_array, ddof=1)  # ddof=1 for sample standard deviation
    sample_size = len(data_array)
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate the t-statistic for the given confidence level and degrees of freedom
    degrees_of_freedom = sample_size - 1
    t_critical = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

    # Calculate the margin of error
    margin_of_error = t_critical * standard_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return (lower_bound, upper_bound)

def interpret_confidence_interval(data, confidence_level=0.95):
    """
    Calculates and interprets a confidence interval for a dataset.

    Args:
        data (list or numpy array): The dataset.
        confidence_level (float): The desired confidence level.
    """
    lower, upper = calculate_confidence_interval(data, confidence_level)

    print(f"{confidence_level * 100:.0f}% Confidence Interval: ({lower:.4f}, {upper:.4f})")
    print(f"Interpretation: We are {confidence_level * 100:.0f}% confident that the true population mean lies within the interval ({lower:.4f}, {upper:.4f}).")

# Example usage:
data = [25, 30, 32, 28, 35, 29, 31, 33, 27, 34]
interpret_confidence_interval(data)

data2 = np.random.normal(50, 10, 50)
interpret_confidence_interval(data2, confidence_level = 0.99)
```

```python
#18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def plot_normal_pdf(mean, std_dev, x_range=(-4, 4), num_points=1000):
    """
    Calculates and plots the PDF of a normal distribution.

    Args:
        mean (float): The mean of the normal distribution.
        std_dev (float): The standard deviation of the normal distribution.
        x_range (tuple): The range of x-values to plot.
        num_points (int): The number of points to generate for the plot.
    """

    if std_dev <= 0:
        raise ValueError("Standard deviation must be positive.")

    x = np.linspace(x_range[0], x_range[1], num_points)
    pdf = stats.norm.pdf(x, mean, std_dev)

    plt.plot(x, pdf)
    plt.xlabel("Values")
    plt.ylabel("Probability Density")
    plt.title(f"Normal Distribution PDF (μ={mean}, σ={std_dev})")
    plt.grid(True)
    plt.show()

# Example usage:
plot_normal_pdf(mean=0, std_dev=1)  # Standard normal distribution
plot_normal_pdf(mean=5, std_dev=2, x_range=(-1, 11)) # change the x range.
plot_normal_pdf(mean=-2, std_dev=0.5, num_points = 500) # change the number of points.
```

```python
#19.  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def calculate_poisson_cdf(lam, values):
    """
    Calculates the CDF of a Poisson distribution for given values.

    Args:
        lam (float): The lambda (mean) parameter of the Poisson distribution.
        values (list or numpy array): The values to calculate the CDF for.

    Returns:
        numpy array: An array of CDF values corresponding to the given values.
    """

    if lam <= 0:
        raise ValueError("Lambda (lam) must be a positive number.")

    return stats.poisson.cdf(values, lam)

def visualize_poisson_cdf(lam, max_value=10):
    """
    Calculates and visualizes the CDF of a Poisson distribution.

    Args:
        lam (float): The lambda (mean) parameter of the Poisson distribution.
        max_value (int): The maximum value to plot the CDF up to.
    """

    if lam <= 0:
        raise ValueError("Lambda (lam) must be a positive number.")

    x = np.arange(0, max_value + 1)
    cdf = calculate_poisson_cdf(lam, x)

    plt.step(x, cdf, where='post')
    plt.xlabel("Number of Events")
    plt.ylabel("Cumulative Probability")
    plt.title(f"Poisson Distribution CDF (λ={lam})")
    plt.grid(True)
    plt.show()

def interpret_poisson_cdf(lam, value):
    """
    Calculates and interprets the CDF of a Poisson distribution at a specific value.

    Args:
        lam (float): The lambda (mean) parameter of the Poisson distribution.
        value (int): The value to calculate and interpret the CDF for.
    """

    cdf_value = calculate_poisson_cdf(lam, value)

    print(f"CDF at x = {value}: {cdf_value:.4f}")
    print(f"Interpretation: The probability of observing {value} or fewer events is approximately {cdf_value:.4f}.")

# Example usage:
lam = 3.5
values = [0, 1, 2, 3, 4, 5]

cdf_values = calculate_poisson_cdf(lam, values)
print("CDF Values:", cdf_values)

visualize_poisson_cdf(lam)

interpret_poisson_cdf(lam, 3)
interpret_poisson_cdf(lam, 6)
```

```python
#20.  Simulate a random variable using a continuous uniform distribution and calculate its expected value.

import numpy as np

def simulate_continuous_uniform(a, b, num_samples=10000):
    """
    Simulates a random variable from a continuous uniform distribution and calculates its expected value.

    Args:
        a (float): The lower bound of the distribution.
        b (float): The upper bound of the distribution.
        num_samples (int): The number of samples to generate.

    Returns:
        tuple: A tuple containing the simulated samples and the calculated expected value.
    """

    if a >= b:
        raise ValueError("a must be less than b.")

    samples = np.random.uniform(a, b, num_samples)
    expected_value = (a + b) / 2.0  # Formula for expected value of continuous uniform distribution

    return samples, expected_value

# Example usage:
a = 10
b = 20
samples, expected_value = simulate_continuous_uniform(a, b)

print(f"Simulated Samples (first 10): {samples[:10]}")
print(f"Calculated Expected Value: {expected_value}")
print(f"Empirical Mean of Simulated Samples: {np.mean(samples)}") #Shows how the empirical mean approaches the expected value.
```

```python
#21. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import matplotlib.pyplot as plt
import numpy as np

def compare_std_dev(data1, data2):
    """
    Compares the standard deviations of two datasets and visualizes the difference.

    Args:
        data1 (list or numpy array): The first dataset.
        data2 (list or numpy array): The second dataset.
    """

    if not isinstance(data1, (list, np.ndarray)) or not isinstance(data2, (list, np.ndarray)):
        raise ValueError("Data must be a list or numpy array.")

    if len(data1) < 2 or len(data2) < 2:
        raise ValueError("Data must contain at least two elements to calculate standard deviation.")

    std_dev1 = np.std(data1, ddof=1)  # Sample standard deviation (ddof=1)
    std_dev2 = np.std(data2, ddof=1)

    print(f"Standard Deviation of Data 1: {std_dev1:.4f}")
    print(f"Standard Deviation of Data 2: {std_dev2:.4f}")
    print(f"Difference in Standard Deviations: {std_dev1 - std_dev2:.4f}")

    # Visualization
    labels = ['Data 1', 'Data 2']
    std_devs = [std_dev1, std_dev2]

    plt.bar(labels, std_devs, color=['skyblue', 'salmon'])
    plt.ylabel("Standard Deviation")
    plt.title("Comparison of Standard Deviations")
    plt.show()

# Example usage:
data1 = [10, 12, 15, 18, 20, 22, 25, 30]
data2 = [15, 16, 17, 18, 19, 20, 21, 22]

compare_std_dev(data1, data2)

data3 = np.random.normal(50, 10, 100)
data4 = np.random.normal(50, 5, 100)

compare_std_dev(data3, data4)

#Example of invalid input.
#compare_std_dev([1],[2])
#compare_std_dev(1,2)
```

```python
#22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np
import scipy.stats as stats

def calculate_range_iqr(data):
    """
    Calculates the range and interquartile range (IQR) of a dataset.

    Args:
        data (list or numpy array): The dataset.

    Returns:
        tuple: A tuple containing the range and IQR.
    """

    if not isinstance(data, (list, np.ndarray)):
        raise ValueError("Data must be a list or numpy array.")

    if len(data) < 2:
        raise ValueError("Data must contain at least two elements to calculate range and IQR.")

    data_array = np.array(data)

    # Calculate range
    data_range = np.max(data_array) - np.min(data_array)

    # Calculate IQR
    q75, q25 = np.percentile(data_array, [75, 25])
    iqr = q75 - q25

    return data_range, iqr

def generate_and_calculate_range_iqr(mean, std_dev, sample_size=100):
    """
    Generates data from a normal distribution and calculates its range and IQR.

    Args:
        mean (float): The mean of the normal distribution.
        std_dev (float): The standard deviation of the normal distribution.
        sample_size (int): The number of samples to generate.

    Returns:
        tuple: A tuple containing the range and IQR.
    """

    if std_dev <= 0:
        raise ValueError("Standard deviation must be positive.")

    data = np.random.normal(mean, std_dev, sample_size)
    data_range, iqr = calculate_range_iqr(data)

    print(f"Generated Data (first 10): {data[:10]}")
    print(f"Range: {data_range:.4f}")
    print(f"IQR: {iqr:.4f}")

    return data_range, iqr

# Example usage:
generate_and_calculate_range_iqr(mean=50, std_dev=10)
generate_and_calculate_range_iqr(mean=100, std_dev=15, sample_size=50)

#Example of invalid inputs.
#generate_and_calculate_range_iqr(1, 0)
#calculate_range_iqr([1])
```

```python
#23. Implement Z-score normalization on a dataset and visualize its transformation.

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def z_score_normalize(data):
    """
    Performs Z-score normalization on a dataset.

    Args:
        data (list or numpy array): The dataset.

    Returns:
        numpy array: The normalized dataset.
    """

    if not isinstance(data, (list, np.ndarray)):
        raise ValueError("Data must be a list or numpy array.")

    if len(data) < 2:
        raise ValueError("Data must contain at least two elements to calculate Z-scores.")

    data_array = np.array(data)
    normalized_data = stats.zscore(data_array)
    return normalized_data

def visualize_zscore_normalization(data):
    """
    Visualizes the transformation of a dataset after Z-score normalization.

    Args:
        data (list or numpy array): The original dataset.
    """

    normalized_data = z_score_normalize(data)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.hist(data, bins=20, alpha=0.6, color='skyblue', edgecolor='black')
    plt.title("Original Data")
    plt.xlabel("Values")
    plt.ylabel("Frequency")

    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=20, alpha=0.6, color='salmon', edgecolor='black')
    plt.title("Normalized Data (Z-scores)")
    plt.xlabel("Z-scores")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

# Example usage:
data = np.random.normal(50, 20, 1000)  # Example dataset from a normal distribution
visualize_zscore_normalization(data)

data2 = [1, 2, 3, 4, 5, 10, 20, 30] #Example with outliers.
visualize_zscore_normalization(data2)

#Example of invalid inputs.
#visualize_zscore_normalization([1])
#visualize_zscore_normalization(1)
```

```python
#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis(data):
    """
    Calculates the skewness and kurtosis of a dataset.

    Args:
        data (list or numpy array): The dataset.

    Returns:
        tuple: A tuple containing the skewness and kurtosis.
    """

    if not isinstance(data, (list, np.ndarray)):
        raise ValueError("Data must be a list or numpy array.")

    if len(data) < 4:  # Skewness and kurtosis require at least 4 data points
        raise ValueError("Data must contain at least four elements to calculate skewness and kurtosis.")

    data_array = np.array(data)
    skewness = stats.skew(data_array)
    kurtosis = stats.kurtosis(data_array)

    return skewness, kurtosis

def generate_and_calculate_skewness_kurtosis(mean, std_dev, sample_size=100):
    """
    Generates data from a normal distribution and calculates its skewness and kurtosis.

    Args:
        mean (float): The mean of the normal distribution.
        std_dev (float): The standard deviation of the normal distribution.
        sample_size (int): The number of samples to generate.

    Returns:
        tuple: A tuple containing the skewness and kurtosis.
    """

    if std_dev <= 0:
        raise ValueError("Standard deviation must be positive.")

    data = np.random.normal(mean, std_dev, sample_size)
    skewness, kurtosis = calculate_skewness_kurtosis(data)

    print(f"Generated Data (first 10): {data[:10]}")
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis: {kurtosis:.4f}")

    return skewness, kurtosis

# Example usage:
generate_and_calculate_skewness_kurtosis(mean=0, std_dev=1)
generate_and_calculate_skewness_kurtosis(mean=5, std_dev=2, sample_size=500)

#Example of invalid inputs.
#generate_and_calculate_skewness_kurtosis(1,0)
#calculate_skewness_kurtosis([1,2,3])
```
